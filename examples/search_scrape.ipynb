```python
import play_scraper
import pandas as pd
```

```python
alphabet = ['a','b','c','d','e','f']
list_len = len(alphabet)

app_list = []

for i in range(list_len):
    search_term = alphabet[i]
    for j in range(list_len):
        search_term = search_term + alphabet[j]
        print('Searching APPs with:', search_term)
        
        for page in range(12):
            tmp_list = play_scraper.search(search_term, detailed=True, page=page)
            app_list.extend(tmp_list)
            
        search_term = alphabet[i]
        print('collected: ', len(app_list))
        
       
    
# len(play_scraper.search('aa', page=1))
```

```python
import string
alphabet = string.ascii_lowercase
print(alphabet)
search_term_list = []

for i in range(len(alphabet)):
    search_term = alphabet[i]
    for j in range(len(alphabet)):
        search_term = search_term + alphabet[j]
        search_term_list.append(search_term)
        search_term = alphabet[i]
        

with open('search_term_list.txt', 'w') as f:
    for item in search_term_list:
        f.write("%s\n" % item)
        
```
